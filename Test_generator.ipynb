## Tokenize and Preprocess Text

### Subtask:
Implement a function to preprocess the text by lowercasing, extracting words using `RegexpTokenizer`, and filtering out stopwords. The result will be a cleaned sequence of characters.


**Reasoning**:
I will implement the text preprocessing function as described, including lowercasing, tokenizing with `RegexpTokenizer`, filtering stopwords, and joining the processed words into a single string. This directly addresses the subtask instructions.



```python
import re
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords

# 1. Convert the entire text_data to lowercase.
text_data_lower = text_data.lower()

# 2. Initialize a RegexpTokenizer to extract words.
tokenizer = RegexpTokenizer(r'\w+')

# 3. Get the list of English stopwords.
english_stopwords = set(stopwords.words('english'))

# 4. Define a function to tokenize and filter stopwords.
def preprocess_text(text):
    tokens = tokenizer.tokenize(text)
    filtered_tokens = [word for word in tokens if word not in english_stopwords]
    return filtered_tokens

# 5. Apply this preprocessing function to your lowercase text_data.
cleaned_words = preprocess_text(text_data_lower)

# 6. Join the cleaned words back into a single string.
processed_text_data = " ".join(cleaned_words)

print(f"Original text data length: {len(text_data)}")
print(f"Lowercased text data length: {len(text_data_lower)}")
print(f"Number of cleaned words: {len(cleaned_words)}")
print(f"Processed text data length (joined string): {len(processed_text_data)}")
print("\nFirst 500 characters of the processed text data:")
print(processed_text_data[:500] + "...")
```

## Create Character Mappings

### Subtask:
Generate unique sorted characters from the preprocessed text. Create two dictionaries: one to map characters to numerical IDs (`char_to_num`) and another to map numerical IDs back to characters (`num_to_char`). Display the total number of characters and unique vocabulary size.


**Reasoning**:
The subtask requires generating unique sorted characters, creating character-to-ID and ID-to-character mappings, and displaying the total number of characters and the vocabulary size. This can be achieved with a single Python code block.



```python
import numpy as np

# 1. Get all unique characters from the processed_text_data string and convert to a sorted list.
# Note: processed_text_data is a string of words separated by spaces.
# We need to consider each character in this string as a potential token if we are building a character-level model.
# However, the previous tokenizer was WordPiece, which suggests word/subword level.
# The subtask specifically asks for 'characters', so we'll process the 'processed_text_data' string character by character.

# Ensure processed_text_data is treated as a sequence of characters for vocabulary building
all_characters = sorted(list(set(processed_text_data)))

# 2. Create a dictionary named char_to_num that maps each unique character to a numerical ID.
char_to_num = {char: i for i, char in enumerate(all_characters)}

# 3. Create a dictionary named num_to_char that maps each numerical ID back to its corresponding character.
num_to_char = np.array(all_characters)

# 4. Print the total number of characters in the processed_text_data.
print(f"Total characters in processed text data: {len(processed_text_data)}")

# 5. Print the size of the unique character vocabulary.
vocabulary_size_char = len(all_characters)
print(f"Unique character vocabulary size: {vocabulary_size_char}")

print("\nFirst 10 char_to_num mappings:")
for i, (char, num) in enumerate(char_to_num.items()):
    if i >= 10: break
    print(f"  '{char}': {num}")

print("\nFirst 10 num_to_char mappings (from array):")
for i in range(min(10, len(num_to_char))):
    print(f"  {i}: '{num_to_char[i]}' ")
```

## Prepare Training Sequences

### Subtask:
Divide the numericalized text into input sequences and corresponding next-character outputs. Each input sequence will have a defined length (`seq_length`), and the output will be the character that immediately follows it. Reshape the input data to be compatible with LSTM layers and normalize the values. One-hot encode the target output characters.


**Reasoning**:
I will convert the `processed_text_data` string into numerical IDs, create input-output sequences based on `SEQ_LEN`, reshape and normalize the input sequences, and one-hot encode the target characters, as per the instructions.



```python
import numpy as np
import tensorflow as tf

# 1. Convert the processed_text_data string into a list of numerical IDs
numerical_text = [char_to_num[char] for char in processed_text_data]

# 2. Initialize empty lists
X = [] # for input sequences
y = [] # for target characters

# 3. Iterate through the numerical_text to create sequences and targets
# The loop should stop SEQ_LEN characters before the end to avoid index errors
for i in range(len(numerical_text) - SEQ_LEN):
    sequence_in = numerical_text[i:i + SEQ_LEN]
    sequence_out = numerical_text[i + SEQ_LEN]
    X.append(sequence_in)
    y.append(sequence_out)

# 4. Convert X and y lists into NumPy arrays
X = np.array(X)
y = np.array(y)

# 5. Reshape the X array for LSTM layer: (number_of_sequences, SEQ_LEN, 1)
# The third dimension is 1 because each character is a single numerical ID
X = X.reshape(X.shape[0], SEQ_LEN, 1)

# 6. Normalize the input sequences in X by dividing by vocabulary_size_char
X = X / float(vocabulary_size_char)

# 7. One-hot encode the target characters in y
y = tf.keras.utils.to_categorical(y, num_classes=vocabulary_size_char)

print(f"Shape of input sequences (X): {X.shape}")
print(f"Shape of target characters (y): {y.shape}")
print(f"First 5 input sequences (numerical, normalized):\n{X[:5].squeeze()}")
print(f"First 5 target characters (one-hot encoded):\n{y[:5]}")
```

## Build LSTM Model

### Subtask:
Construct a Keras Sequential model using LSTM layers for sequence processing and Dropout layers for regularization. The model architecture will consist of multiple LSTM layers, followed by a Dense output layer with a 'softmax' activation function, tailored for character prediction.


**Reasoning**:
I will construct the Keras Sequential model as specified, incorporating LSTM and Dropout layers, and then print a summary of the model architecture to verify its structure.



```python
from keras.models import Sequential
from keras.layers import LSTM, Dropout, Dense

# 1. Initialize a Sequential model.
model = Sequential()

# 2. Add the first LSTM layer
# input_shape is (SEQ_LEN, 1) based on X.shape after reshaping
model.add(LSTM(256, input_shape=(X.shape[1], X.shape[2]), return_sequences=True))

# 3. Add a Dropout layer
model.add(Dropout(0.2))

# 4. Add a second LSTM layer
model.add(LSTM(256, return_sequences=True))

# 5. Add another Dropout layer
model.add(Dropout(0.2))

# 6. Add a third LSTM layer
model.add(LSTM(128)) # return_sequences is False by default here

# 7. Add a final Dense output layer
model.add(Dense(vocabulary_size_char, activation='softmax'))

# Print the model summary to review the architecture
print(model.summary())
print("Keras LSTM model constructed successfully.")
```

**Reasoning**:
The previous code generated a `UserWarning` from Keras about `input_shape`. To resolve this and follow Keras's best practices for Sequential models, I will explicitly add an `Input` layer as the first layer, which is the recommended approach for defining input shape.



```python
from keras.models import Sequential
from keras.layers import LSTM, Dropout, Dense, Input

# 1. Initialize a Sequential model.
model = Sequential()

# Add an Input layer explicitly to avoid UserWarning about input_shape in Sequential models
model.add(Input(shape=(X.shape[1], X.shape[2])))

# 2. Add the first LSTM layer
model.add(LSTM(256, return_sequences=True))

# 3. Add a Dropout layer
model.add(Dropout(0.2))

# 4. Add a second LSTM layer
model.add(LSTM(256, return_sequences=True))

# 5. Add another Dropout layer
model.add(Dropout(0.2))

# 6. Add a third LSTM layer
model.add(LSTM(128)) # return_sequences is False by default here

# 7. Add a final Dense output layer
model.add(Dense(vocabulary_size_char, activation='softmax'))

# Print the model summary to review the architecture
print(model.summary())
print("Keras LSTM model constructed successfully.")
```

```python
from keras.callbacks import ModelCheckpoint

# 1. Compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
print("Model compiled successfully.")

# 2. Define a filepath for saving the model weights
filepath = "weights-improvement-{epoch:02d}-{loss:.4f}.keras"

# 3. Create a ModelCheckpoint callback
checkpoint = ModelCheckpoint(
    filepath,
    monitor='loss',
    verbose=1,
    save_best_only=True,
    mode='min'
)

callbacks_list = [checkpoint]

# 4. Train the model
print("Starting model training...")
history = model.fit(
    X,
    y,
    epochs=20, # A reasonable starting number of epochs
    batch_size=128,
    callbacks=callbacks_list
)

print("Model training completed.")
```
